```python
y_train_pred_prob=np.array(y_train_pred_prob)
```

```python
results_df = pd.DataFrame({
    'Actual_Stayed': y_train_encoded.values.squeeze(),
    'Predicted_Prob_Stayed': y_train_pred_prob
})
```

```python
results_df['Predicted']=results_df['Predicted_Prob_Stayed'].apply(lambda x: 1 if x >0.52 else 0)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_curve, roc_auc_score, roc_curve, recall_score,precision_score
accuracy = accuracy_score(results_df['Actual_Stayed'],results_df['Predicted'])
accuracy
```

```python
print(precision_score(results_df['Actual_Stayed'],results_df['Predicted']))
print(recall_score(results_df['Actual_Stayed'],results_df['Predicted']))
```

```python
for i in np.arange(0, 1, 0.1):
    threshold = round(i, 1)
    results_df[f'predicted_{threshold:.2f}'] = results_df['Predicted_Prob_Stayed'].apply(lambda x: 1 if x > threshold else 0)
```

```python
from sklearn.metrics import confusion_matrix

# Initialize list to store metrics
cutoff_metrics = []

# Loop through all cutoff columns
for cutoff in np.arange(0, 1, 0.1):
    cutoff = round(cutoff, 1)
    y_pred = results_df[f'predicted_{cutoff:.2f}']
    y_actual = results_df['Actual_Stayed']

    tn, fp, fn, tp = confusion_matrix(y_actual, y_pred).ravel()

    accuracy = (tp + tn) / (tp + tn + fp + fn)
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)

    cutoff_metrics.append({
        'Cutoff': cutoff,
        'Accuracy': round(accuracy, 3),
        'Sensitivity': round(sensitivity, 3),
        'Specificity': round(specificity, 3)
    })
'''built a cutoff-wise evaluation DataFrame to compare model performance metrics and found the best trade-off near a 0.5 probability threshold.'''
# Create DataFrame
cutoff_df = pd.DataFrame(cutoff_metrics)
cutoff_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(cutoff_df['Cutoff'], cutoff_df['Accuracy'], label='Accuracy', marker='o')
plt.plot(cutoff_df['Cutoff'], cutoff_df['Sensitivity'], label='Sensitivity (Recall)', marker='o')
plt.plot(cutoff_df['Cutoff'], cutoff_df['Specificity'], label='Specificity', marker='o')

plt.xlabel('Probability Cutoff')
plt.ylabel('Score')
plt.title('Accuracy, Sensitivity, and Specificity at Different Cutoffs')
plt.legend()
plt.grid(True)
plt.ylim(0, 1)
plt.xlim(0, 0.9)

plt.show()
```
